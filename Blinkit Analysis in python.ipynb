## **BUSINESS REQUIREMENTS**


# **KPI Requirements**
**Total Sales**: The overall revenue generated from all items sold.

**Average Sales**: The average revenue per sale.

**Number of Items**: The total count of different items sold.

**Average Rating**: The average customer rating for items sold.


### **KPI's REQUIREMENTS**

```python
#Total Sales
total_sales = df['Sales'].sum()

#Average sales
avg_sales = df['Sales'].mean()

#Number of items sold
no_of_items_sold = df['Sales'].count()

#Average ratings
average_ratings = df['Rating'].mean()

#Display
print(f"Total Sales: ${total_sales:,.0f}")  #formated string  #, thousands seperaeted 
print(f"Average sales: ${avg_sales:,.0f}")
print(f"No of item sold: ${no_of_items_sold:,.0f}")
print(f"average_ratings: ${average_ratings:,.0f}")
```

### **CHARTs REQUIREMENTS**

#### **Total Sales by Fat Content**

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

plt.pie(sales_by_fat, labels = sales_by_fat.index,
         autopct = '%.1f%%',
                      startangle = 90)

plt.title('Sales by Fat Content')
plt.axis('equal')
plt.show()
```

```python
## Bar Graph with data call-out values and resized graph

sales_by_type = df.groupby( 'Item Type') ['Sales'].sum().sort_values(ascending=False)
plt. figure(figsize=(10, 6))
bars = plt.bar(sales_by_type.index, sales_by_type.values)
plt.xticks(rotation=-90)
plt. xlabel( 'Item Type')
plt. ylabel('Total Sales')
plt. title('Total Sales by Item Type')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'{bar.get_height():,.0f}', ha='center', va= 'bottom', fontsize=8)
plt. tight_layout()
plt. show()
```

### **Fat Content by Outlet for Total Sales**

```python
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()
grouped = grouped[['Regular', 'Low Fat']]

ax = grouped.plot(kind='bar', figsize=(8, 5), title='Oulet Tier by Item Fat Content')
plt.xlabel('Outlet Location Tier')
plt.ylabel('Total sales')
plt.legend(title='Item Fat Content')
plt.tight_layout()
plt.show()
```

### **Total Sales by Outlet Establishment**

```python
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()
plt.figure(figsize = (10,6))
plt.plot(sales_by_year.index, sales_by_year.values , marker = 's', linestyle= '--')
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Outlet Establishment')

for x, y in zip(sales_by_year.index, sales_by_year.values+5):
    plt.text(x, y + 1000, f'{int(y):,}', ha='center', va='bottom', fontsize=9)
plt.show()
```

### **Total Sales by Outlet Size**

```python
sales_by_size= df.groupby('Outlet Size')['Sales'].sum()

plt.pie(sales_by_size,labels =sales_by_size.index , autopct = '%1.0f%%')
plt.title('Sales by Outlet Size')
plt.tight_layout()
plt.show() 
```

### **Total Sales by Outlet Location**

```python
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales', ascending=False)

plt.figure(figsize=(8,5)) # smaller height, enough width
ax = sns.barplot(x='Sales', y='Outlet Location Type', data= sales_by_location)

plt.title('Total sals by Outlet Location Type')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')

plt.tight_layout()  #Ensures layout fits without scroll
plt.show()
```

## **Item Weight vs Sales by Fat Content**

```python
df_clean = df.dropna(subset=["Item Weight"])
sns.set(style="whitegrid")
# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_clean,
    x="Item Weight",
    y="Sales",
    hue="Item Fat Content",
    alpha=0.6,
    palette="Set2"
)

# Title and axis labels
plt.title("Item Weight vs Sales by Fat Content", fontsize=14)
plt.xlabel("Item Weight")
plt.ylabel("Sales")
plt.tight_layout()

# Show the plot
plt.show()

```

```python

# Set the visual style
sns.set(style="whitegrid")

# Group by Item Type and calculate average sales
avg_sales = df.groupby("Item Type")["Sales"].mean().sort_values()

# Create the bar plot
plt.figure(figsize=(9, 5))
sns.barplot(x=avg_sales.values, y=avg_sales.index, palette="viridis")

# Add chart labels and title
plt.title("Average Sales by Item Type", fontsize=14)
plt.xlabel("Average Sales")
plt.ylabel("Item Type")
plt.tight_layout()

# Show the plot
plt.show()
```
